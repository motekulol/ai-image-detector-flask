Pliki w zestawie Lite Dataset:
1. photos.tsv000
   - ID zdjęcia (photo_id).
   - Linki do zdjęć (photo_image_url) – to kluczowe, aby pobrać zdjęcia.
   - Metadane (photo_width, photo_height, exif_camera_model, ai_description, stats_views, stats_downloads) – pomocne do filtrowania zdjęć.
   - Podstawowy plik do pracy, ponieważ zawiera linki i główne informacje o zdjęciach.
   - Posłuży do filtrowania zdjęć na podstawie rozdzielczości, opisu lub liczby pobrań.
3. collections.tsv000
   - ID kolekcji (collection_id) i jej nazwę (title).
   - Opis kolekcji.
   - Słowa kluczowe związane z kolekcją.
   - ID zdjęć należących do kolekcji (photo_ids).
   - Jeśli chcesz wybrać zdjęcia związane z określoną kolekcją (np. „architecture”, „nature”).
   - Można połączyć to z plikiem photos.tsv000, aby filtrować zdjęcia wg kolekcji.
5. keywords.tsv000
   - ID zdjęcia (photo_id).
   - Słowa kluczowe powiązane z danym zdjęciem.
   - Przydatne do bardziej precyzyjnego filtrowania zdjęć na podstawie kategorii.
   - Można połączyć z photos.tsv000 na podstawie kolumny photo_id, aby dodać dodatkowy kontekst do zdjęć.
7. colors.tsv000
   - ID zdjęcia (photo_id).
   - Dominujące kolory w formacie HEX.
   - Może być interesujące, jeśli Twój projekt wymaga analizy kolorów zdjęć (np. w kontekście rozróżniania zdjęć AI od rzeczywistych).
   - Nie jest kluczowe, ale może stanowić dodatkowy kontekst.
9. conversions.tsv000
   - ID zdjęcia (photo_id).
   - Liczba kliknięć, pobrań, interakcji.
   - Można użyć, aby wybrać popularne zdjęcia o wysokiej liczbie pobrań lub wyświetleń.
   - Przydatne, jeśli chcesz analizować zdjęcia, które są często używane w praktyce.y pobrań.kolekcji.y pobrań.

```python
categories = ["building", "food", "animal"]
base_folder = "C:\\Users\\zonea\\Desktop\\WSB\\seminarium\\Obrazy"
os.makedirs(base_folder, exist_ok=True)

# Lista słów kluczowych wskazujących na rendery 3D
render_keywords = ["3D", "render", "illustration", "CGI", "computer generated"]

# Ustawienia przetwarzania obrazów
target_size = (512, 512)  # Docelowy rozmiar obrazów
```

```python
for category in categories:
    # Filtrowanie danych dla kategorii
    category_df = merged_df[
        (merged_df['keyword'].str.contains(category, na=False)) &
        (merged_df['ai_service_1_confidence'] >= 90) & 
        (~merged_df['ai_description'].str.contains('|'.join(render_keywords), na=False, case=False)) &
        (~merged_df['keyword'].str.contains('|'.join(render_keywords), na=False, case=False)) &
        (merged_df['exif_camera_model'].notna())
    ]

    # Usuwanie duplikatów na podstawie photo_id
    category_df = category_df.drop_duplicates(subset="photo_id").head(250)
    
    # Tworzenie folderu dla kategorii
    category_folder = os.path.join(base_folder, category)
    os.makedirs(category_folder, exist_ok=True)

    # Licznik do numeracji zdjęć
    counter = 1

    # Pobieranie zdjęć
    for _, row in category_df.iterrows():
        photo_id = row['photo_id']
        image_url = row['photo_image_url']
        try:
            # Pobranie obrazu
            response = requests.get(image_url)
            if response.status_code == 200:
                # Otwórz obraz, zmień rozmiar i konwertuj na RGB
                image = Image.open(BytesIO(response.content)).convert("RGB")
                image = image.resize(target_size)
                
                # Zapisz obraz z odpowiednią nazwą
                save_path = os.path.join(category_folder, f"real_{category}_{counter:03}.jpeg")
                image.save(save_path, "JPEG")
                counter += 1
            else:
                print(f"Nie udało się pobrać zdjęcia: {photo_id}")
        except Exception as e:
            print(f"Błąd przy pobieraniu {photo_id}: {e}")
```

```python
categories = ["building", "food", "animal"]
base_folder = "C:\\Users\\zonea\\Desktop\\WSB\\seminarium\\Obrazy"

# Lista słów kluczowych wskazujących na rendery 3D
render_keywords = ["3D", "render", "illustration", "CGI", "computer generated"]

# Dane do zapisania opisów
all_descriptions = []

for category in categories:
    # Filtrowanie danych dla kategorii
    category_df = merged_df[
        (merged_df['keyword'].str.contains(category, na=False)) &
        (merged_df['ai_service_1_confidence'] >= 90) & 
        (~merged_df['ai_description'].str.contains('|'.join(render_keywords), na=False, case=False)) &
        (~merged_df['keyword'].str.contains('|'.join(render_keywords), na=False, case=False)) &
        (merged_df['exif_camera_model'].notna())
    ]

    # Usuwanie duplikatów na podstawie photo_id
    category_df = category_df.drop_duplicates(subset="photo_id").head(250)

    # Przechowywanie opisów
    for _, row in category_df.iterrows():
        photo_id = row['photo_id']
        all_descriptions.append({
            "category": category,
            "photo_id": photo_id,
            "photo_description": row.get("photo_description", ""),
            "ai_description": row.get("ai_description", ""),
            "confidence": row.get("ai_service_1_confidence", "")
        })

# Zapisanie opisów do pliku CSV
descriptions_df = pd.DataFrame(all_descriptions)
descriptions_df.to_csv(os.path.join(base_folder, "photo_descriptions.csv"), index=False, encoding="utf-8")

print("Opisy zostały zapisane do 'photo_descriptions.csv'.")
```

```python
base_folder = "C:\\Users\\zonea\\Desktop\\WSB\\seminarium\\Obrazy"
descriptions_path = os.path.join(base_folder, "photo_descriptions.csv")
descriptions_df = pd.read_csv(descriptions_path)

# Funkcja generująca prompt na podstawie opisów
def generate_prompt(row):
    if pd.notna(row["photo_description"]) and row["photo_description"].strip():
        description = row["photo_description"]
    else:
        description = row["ai_description"]
    
    # Dodanie dodatkowych elementów do promptu
    prompt = f"{description}, high resolution, professional photography"
    return prompt

# Tworzenie promptów
descriptions_df["prompt"] = descriptions_df.apply(generate_prompt, axis=1)

# Zapisanie promptów do pliku CSV
prompts_path = os.path.join(base_folder, "ai_prompts.csv")
descriptions_df[["category", "photo_id", "prompt"]].to_csv(prompts_path, index=False, encoding="utf-8")

print(f"Prompty zostały zapisane do pliku '{prompts_path}'.")
```
